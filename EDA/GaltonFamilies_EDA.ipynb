```python
import pandas as pd

def find_outliers_iqr(df):
    outliers_list = []
    for column in df.select_dtypes(include=[float, int]).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        outlier_mask = (df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))
        outliers_list.append(df[outlier_mask])
    outliers = pd.concat(outliers_list).drop_duplicates()
    return outliers



# Find outliers using IQR method
iqr_outliers = find_outliers_iqr(Data)
print("Outliers based on IQR:\n", iqr_outliers)
```
